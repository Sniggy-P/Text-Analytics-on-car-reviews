```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import datetime as dt
import pandas as pd
```

```python
def review_text():
    r_text=driver.find_elements_by_tag_name('div.review.js-more')
    r_list=[i.text for i in r_text]
    r_series=pd.Series(r_list)
    return r_series
```

```python
def author():
    a_text=driver.find_elements_by_class_name('author')
    a_list=[i.text for i in a_text]
    a_series=pd.Series(a_list)
    return a_series
```

```python
def r_date():
    d_text=driver.find_elements_by_tag_name('span.date.icon.icon-calendar-2')
    #d_list=[i.text for i in d_text]
    d_list=[dt.datetime.strptime(i.text,'%b %d, %Y %H:%M %p').strftime('%Y-%m-%d') for i in d_text]
    d_series=pd.Series(d_list)
    return d_series
```

```python
def r_star():
    s_text=driver.find_elements_by_xpath(
    "/html/body/div[3]/div[1]/div[2]/div/div[3]/div/div/div/div[1]/div/div/ul/li/div[3]/div[1]/div[1]/span/span")
    s_list=[i.get_attribute('data-rating') for i in s_text]
    s_series=pd.Series(s_list)
    return s_series
```

```python
#Chrome driver
locationOfWebdriver = "C:\\Users\\Varun Sethi\\Documents\\AMPBA\\chromedriver_win32\\chromedriver.exe"
driver = webdriver.Chrome(locationOfWebdriver)
```

```python
brand=['Jeep','Kia','MG']
car=['Compass','Seltos','Hector']
Website='https://auto.ndtv.com'
review_df=pd.DataFrame()
for i in range(0,3):
    driver.get('https://auto.ndtv.com/%s-cars/%s/reviews'%(brand[i],car[i]))
    #time.sleep(5)
    ndtv_df=pd.DataFrame(columns=['review_website','car_brand','car_model','city','date',
                                  'star_rating','review_text'])
    ndtv_df.date=r_date()
    ndtv_df.star_rating=r_star()
    ndtv_df.review_text=review_text()
    ndtv_df.review_website=Website
    ndtv_df.car_brand=brand[i]
    ndtv_df.car_model=car[i]
    review_df=review_df.append(ndtv_df,ignore_index=True)
    #print(Website)
    time.sleep(3)
    
#driver.close()
##dt.datetime.strptime('Dec 06, 2019 07:37 AM','%b %d, %Y %H:%M %p').strftime('%Y-%m-%d')
review_df
```
